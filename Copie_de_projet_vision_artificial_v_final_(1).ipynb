```python
#addition
"""
a. Donner l image ’ , résultante de l opération ’ . Vérifier le résultat obtenu
en le comparaissant avec le résultat mathématique de cette opération.
"""


import numpy as np

matrice1 = []
matrice2 = []

with open('/content/image1_binaire.pbm') as fichier1:
    for ligne in fichier1:
        matrice1.append(ligne.rstrip('\n'))

with open('/content/image2_binaire.pbm') as fichier2:
    for ligne in fichier2:
        matrice2.append(ligne.rstrip('\n'))

entete = matrice1[:2]
entete_concatene = ""

for i in range(2):
    entete_concatene += ''.join(str(el) for el in entete[i])

matrice1 = matrice1[2:]
matrice2 = matrice2[2:]

for ligne_index in range(0, 9):
    ligne = list(matrice1[ligne_index].strip(" ").split())
    for col_index in range(0, 9):
        try:
            ligne[col_index] = int(ligne[col_index])
        except ValueError:
            ligne[col_index] = 0
    matrice1[ligne_index] = ligne

for ligne_index in range(0, 9):
    ligne = matrice2[ligne_index].strip().split()
    for col_index in range(0, 9):
        try:
            if ligne[col_index]:
                ligne[col_index] = int(ligne[col_index])
            else:
                ligne[col_index] = 0
        except ValueError:
            ligne[col_index] = 0
    matrice2[ligne_index] = ligne

matrice1 = np.array(matrice1)
matrice2 = np.array(matrice2)

somme_matrice = matrice1 + matrice2

for i in range(len(somme_matrice)):
    for j in range(len(somme_matrice[i])):
        if somme_matrice[i][j] > 1:
            somme_matrice[i][j] = 1

liste_resultat = somme_matrice.tolist()
texte_resultat = [""] * len(liste_resultat)

for i in range(len(liste_resultat)):
    texte_resultat[i] = ' '.join(str(element) for element in liste_resultat[i])

with open('fusion_image.txt', 'w') as fichier_sortie:
    fichier_sortie.writelines('\n'.join(entete + texte_resultat))
```

```python
"""
2. En utilisant le même langage de programmation utilisé dans la question 1, donner
l histogramme d une image couleur représentant votre portrait
"""

import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/content/mon_image.jpg')

if image is None:
    print("Problème : l'image n'a pas pu être chargée.")
else:
    color_channels = ('b', 'g', 'r')
    plt.figure()
    plt.title("Distribution des couleurs dans l'image")
    plt.xlabel("Valeur d'intensité de la couleur")
    plt.ylabel("Fréquence des pixels")

    for i, col in enumerate(color_channels):
        histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
        plt.plot(histogram, color=col)
        plt.xlim([0, 256])

    plt.legend(color_channels)
    plt.show()

```

```python
"""
3. Ecrire un programme permettant de transformer l image utilisée dans la question 2 en une ’
image en niveau de gris ou binaire (il doit permettre à l utilisateur de choisir entre image ’
binaire ou en niveau de gris et aussi le nombre de bit pour cette dernière).
"""

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

photo = cv2.imread('/content/img2.png')

if photo is None:
    print("Erreur : Impossible de charger l'image.")
else:
    choix_utilisateur = input("Choisissez l'option de transformation :\n1 - Conversion en niveaux de gris\n2 - Transformation en image binaire\nSélectionnez : ")

    if choix_utilisateur == '1':
        photo_grise = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)

        nb_bits = int(input("Définissez le nombre de bits pour l'image en niveaux de gris (entre 1 et 8) : "))

        if 1 <= nb_bits <= 8:
            niveaux_de_gris = 2 ** nb_bits
            photo_quantifiee = (photo_grise / (256 / niveaux_de_gris)).astype(np.uint8) * (256 // niveaux_de_gris)

            cv2_imshow(photo_quantifiee)
        else:
            print("Erreur : Le nombre de bits doit être entre 1 et 8.")

    elif choix_utilisateur == '2':
        photo_grise = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)

        _, photo_binaire = cv2.threshold(photo_grise, 127, 255, cv2.THRESH_BINARY)

        cv2_imshow(photo_binaire)

    else:
        print("Sélection invalide. Choisissez 1 pour la conversion en niveaux de gris ou 2 pour l'image binaire.")
```

```python
# 5. Ecrire un programme permettant d appliquer le filtre de Nagao à une image en niveaux de gris.

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def filtre_nagao_rapide(photo, taille_noyau=5):
    photo_floutee = cv2.GaussianBlur(photo, (taille_noyau, taille_noyau), 0)
    laplacien = cv2.Laplacian(photo, cv2.CV_64F)
    laplacien = cv2.convertScaleAbs(laplacien)

    filtre_nagao = cv2.addWeighted(photo_floutee, 1, laplacien, -0.5, 0)
    filtre_nagao = np.clip(filtre_nagao, 0, 255).astype(np.uint8)

    return filtre_nagao

photo = cv2.imread('/content/mon_image.jpg', cv2.IMREAD_GRAYSCALE)

if photo is None:
    print("Erreur : Impossible d'ouvrir la photo.")
else:
    image_filtrée = filtre_nagao_rapide(photo)

    cv2_imshow(photo)
    cv2_imshow(image_filtrée)

    cv2.imwrite('photo_filtrée_nagao.jpg', image_filtrée)
```
